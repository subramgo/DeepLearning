## Basic Image Classification Model

```python
import numpy as np
import h5py
import matplotlib.pyplot as plt
import random

hdf5_path = '../data/cifar-10/hdf5/cifar_10.h5'
hdf5_file = h5py.File(hdf5_path, mode = 'r')


data_size = hdf5_file['train_images'].shape[0]
batch_size = 256
nb_class = 10

print("Total Training Images {}".format(data_size))


def get_random_batch(batch_size, nb_class):
    position = random.randint(0, int(data_size/batch_size)-1) * batch_size
    X_batch = hdf5_file['train_images'][position:position + batch_size]
    labels = hdf5_file['train_labels'][position:position + batch_size]
    X_batch = X_batch / 255
    ## Make Y dummies
    labels_one_hot = np.zeros((batch_size, nb_class))
    labels_one_hot[np.arange(batch_size), labels] = 1
    return X_batch, labels_one_hot
    
```

```python
x_train, y_train = get_random_batch(batch_size, nb_class)
x_train.shape[1]
```

## First Network



input --> conv2d (relu) --> maxpooling --> flatten --> Dense --> softmax


```python
from keras import layers
from keras.layers import Input, Conv2D, Activation,ZeroPadding2D,MaxPooling2D, Dense, Flatten
from keras.models import Model


# Model Specification
x_input = Input(x_train.shape[1:])

X = ZeroPadding2D((2,2))(x_input)

X = Conv2D(10, (5,5), strides = (1,1), name = "conv0")(X)
X = Activation('relu')(X)
X = MaxPooling2D((2,2), name="max_pool0")(X)

X = Flatten()(X)
X = Dense(1000, activation ="relu", name ="fc")(X)
predictions = Dense(10, activation ="softmax")(X)



model = Model(inputs = x_input, outputs = predictions, name = "FirstModel")
model.compile(optimizer = "sgd", loss = "categorical_crossentropy", metrics = ["accuracy"])


for i in range(100):
    x_train, y_train = get_random_batch(batch_size, nb_class)
    model.fit(x = x_train, y = y_train ,epochs = 10, batch_size = 32)

model.evaluate(x=x_train, y=y_train, batch_size=32, verbose=1, sample_weight=None, steps=None)
```
